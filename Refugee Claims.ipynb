## Isolating refugee claims that were decided by the IRB

Isolating the cases in which the IRB made decisions on the file, or in other words, eliminating those that were abandoned or withdrawn.

```python
decided = all_claims[(all_claims.explanation == "Positive") | (all_claims.explanation == "Expedited Positive") | (all_claims.explanation == "Neg. No Cred Basis") | (all_claims.explanation == "Negative")]
```

There were 80,568 claims for which the IRB made a decision during this time period.

Countries with the most decided claims

```python
decided.country.value_counts()
```

## Calculating the average approval rate of all claims

```python
outcomes = pd.pivot_table(decided, columns = "explanation", values="IRB_file_number", aggfunc=np.count_nonzero)
```

```python
outcomes.head()
```

```python
outcomes["Total"] = outcomes.sum(axis=1)
```

```python
outcomes.head()
```

```python
outcomes["Tot_Pos"] = outcomes[ "Positive"].add(outcomes["Expedited Positive"], axis=0)
```

```python
outcomes.head()
```

```python
outcomes["Rate_Pos"] = outcomes["Tot_Pos"]/outcomes["Total"]
```

```python
outcomes.head()
```

Overall, 60 per cent of the claims decided by the IRB over this time period were positive.
